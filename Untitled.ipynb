```python
%matplotlib widget
import numpy as np
import math
import matplotlib.pyplot as plt
import jnpy as jp

fig = plt.figure()
ax = fig.add_subplot(projection='3d')

p = jp.Pocket('1ec0.rec.pdb', '1ec0.lig.mol2', box=20, bin=0.4)
ls1 = p.receptor_grid()

colors = ['blue', 'orange', 'cyan', 'magenta', 'black', 'pink', 'yellow', 'silver']

n = len(colors)

# state = p
# ls = [p.find_children()]
# ps = [child.ligand_grid()[0] for child in ls[-1]]
# for i in range(10):
#   ls.append(ls[-1][selection[i]].find_children())
#   ps = [child.ligand_grid()[0] for child in ls[-1]]
  
#   d = np.array(ps)

children = p.find_children()
ls2 = [child.ligand_grid()[0] for child in children]
print(len(children))

children2 = children[10000].find_children()
ls3 = [child.ligand_grid()[0] for child in children2]
print(len(children2))

# children3 = children2[0].find_children()
# ls4 = [child.ligand_grid()[0] for child in children3]
# print(len(children3))

# children4 = children3[0].find_children()
# ls5 = [child.ligand_grid()[0] for child in children4]
# print(len(children4))

# children5 = children4[0].find_children()
# ls6 = [child.ligand_grid()[0] for child in children5]
# print(len(children5))

chs = [children2]
ls = [ls3]
for i in range(n):
  chs.append(chs[-1][0].find_children())
#   children3 = children2[0].find_children()
  ls.append([child.ligand_grid()[0] for child in chs[-1]])
#   ls4 = [child.ligand_grid()[0] for child in children3]
#   print(len(children3))


def foo(ls1, ls2):
  ls1 = set([tuple(i) for i in ls1])
  ls2 = set([tuple(i) for i in ls2])
  return list(ls1-ls2)

d1 = np.array(ls1)
d2 = np.array(ls2)
d3 = np.array(ls3)
# d4 = np.array(foo(ls4,ls3))
# d5 = np.array(foo(ls5,ls4))
# d6 = np.array(foo(ls6,ls5))
ds = []
for i in range(n):
  ds.append(np.array(foo(ls[i], ls[i+1])))

# print(ls2)

ax.scatter(d1.T[0], d1.T[1], d1.T[2], s=400, color='grey',alpha=0.5)
ax.scatter(d2.T[0], d2.T[1], d2.T[2], color='darkgreen', alpha=0.5, s=0.1)
ax.scatter(d3.T[0], d3.T[1], d3.T[2], color='red', alpha=0.5, s=2)
# ax.scatter(d4.T[0], d4.T[1], d4.T[2], color='blue', alpha=0.5, s=2)
# ax.scatter(d5.T[0], d5.T[1], d5.T[2], color='orange', alpha=0.5, s=2)
# ax.scatter(d6.T[0], d6.T[1], d6.T[2], color='cyan', alpha=0.5, s=2)

for i in range(n):
  ax.scatter(ds[i].T[0], ds[i].T[1], ds[i].T[2], color=colors[i], alpha=0.5, s=2)

plt.show()
```
